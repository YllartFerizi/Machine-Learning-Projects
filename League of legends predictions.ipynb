```python
from sklearn.base import BaseEstimator,TransformerMixin
```

```python
class dropdupes(BaseEstimator,TransformerMixin):
    def __init__(self,unique_column):
        self.unique_column = unique_column
    def fit(self,x,y=None):
        return self
    def transform(self,pipeline):
        return pipeline.drop_duplicates(self.unique_column)
```

```python
class dropper(BaseEstimator,TransformerMixin):
    def __init__(self,columns):
        self.columns = columns
    def fit(self,x,y=None):
        return self
    def transform(self,pipeline):
        return pipeline.drop(self.columns,axis=1)
```

```python
class Fieldtransformers(BaseEstimator,TransformerMixin):
    label = None
    def fit(self,x,y=None):
        return self
    def transform(self,X):
        df = pd.DataFrame()
        blue_columns = [x for x in X.columns if x.startswith('blue')]
        red_columns = [x.replace('blue','red') for x in blue_columns]
        for b,r in zip(blue_columns,red_columns):
            if r in X.columns:
                name = b.strip('blue')
                value = X[b] - X[r]
                df[name] = value
        self.label = df['Wins']
        return df.drop('Wins',axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline([
    ('dropdupes',dropdupes('gameId')),
    ('fieldtransform',Fieldtransformers()),
    ('scale',StandardScaler())
])
```

```python
def label(pipeline):
    return pipeline.steps[1][1].label.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(full_df,test_size=0.2)
```

```python
X_train = pipe.fit_transform(train)
y_train = label(pipe)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)


from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(), X_train, y_train)
```

```python
X_test = pipe.transform(test)
y_test = label(pipe)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```
